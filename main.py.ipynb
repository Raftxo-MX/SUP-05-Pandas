Qué país tiene el promedio de edad más alto de mujeres? 

```python
df_mujeres = clientes[clientes["gender"] == "female"]
promedio_edad_por_pais = df_mujeres.groupby('country')['age'].mean()
pais_mas_alto = promedio_edad_por_pais.idxmax()
pais_mas_alto
```

Qué país tiene el promedio de edad más bajo de hombres ? 

```python
df_hombres = clientes[clientes["gender"] == "male"]
promedio_edad_por_paisH = df_hombres.groupby('country')['age'].mean()
pais_mas_altoH = promedio_edad_por_pais.idxmin()
pais_mas_altoH
```
